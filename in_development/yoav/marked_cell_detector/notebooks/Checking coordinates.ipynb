```python
import pandas as pd
det=pd.read_csv('../detections_dk55.csv')
positives=det[det['label']==1]
positives.head()
```

```python
!ls $DIR/sections/*/*.csv
```

```python
secs=[]
for section in [164,172,248]:
    sec=pd.read_csv(DIR+'/sections/%d/%d.csv'%(section,section))
    secs.append(sec)
all_secs=pd.concat(secs)
all_secs
```

```python
import numpy as np
from numpy.linalg import norm
```

```python
for section in [164,172,248]:
    
    print('section=',section)

    pos_sec=positives[positives['section']==section]
    pos_xy=np.array(pos_sec[['col','row']])
    #pos_xy

    sec_man=all_secs[all_secs['Section']==section]
    man_xy=np.array(sec_man[['x','y']])
    #man_xy

    for i in range(pos_xy.shape[0]):
        c=norm(man_xy-pos_xy[i,:],axis=1)
        c=np.sort(c)
        print(c[:2])

```
